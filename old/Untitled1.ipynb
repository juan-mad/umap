```python
import sklearn.datasets
import pandas as pd
import numpy as np
import umap
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set(style='white', context='notebook', rc={'figure.figsize':(14,10)})
```

```python
penguins = pd.read_csv("https://raw.githubusercontent.com/allisonhorst/palmerpenguins/c19a904462482430170bfe2c718775ddb7dbb885/inst/extdata/penguins.csv")
penguins.head()
```

```python
penguins = penguins.dropna()
penguins.species.value_counts()
```

```python
sns.pairplot(penguins.drop("year", axis=1), hue='species');
```

```python
reducer = umap.UMAP()
```

```python
penguin_data = penguins[
    [
        "bill_length_mm",
        "bill_depth_mm",
        "flipper_length_mm",
        "body_mass_g",
    ]
].values
scaled_penguin_data = StandardScaler().fit_transform(penguin_data)
```

```python
scaled_penguin_data
```

```python
embedding = reducer.fit_transform(scaled_penguin_data)
embedding.shape
```

```python
plt.scatter(
    embedding[:, 0],
    embedding[:, 1],
    c=[sns.color_palette()[x] for x in penguins.species.map({"Adelie":0, "Chinstrap":1, "Gentoo":2})])
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Penguin dataset', fontsize=24);
```

```python
num_samples = 20

# make a simple unit circle
theta = np.linspace(0, 2 * np.pi, num_samples)
a, b = 1 * np.cos(theta), 1 * np.sin(theta)

# generate the points
# theta = np.random.rand((num_samples)) * (2 * np.pi)
x, y = np.cos(theta), np.sin(theta)
```

```python
orig_data = np.zeros((x.shape[0], 2))
orig_data[:,0] = x
orig_data[:,1] = y
```

```python
plt.figure(figsize=(7, 6))
plt.scatter(orig_data[:,0], orig_data[:,1], label='Samples', marker="o")
plt.ylim([-1.5, 1.5])
plt.xlim([-1.5, 1.5])
plt.grid()
plt.legend(loc='upper right')
plt.show(block=True)
```

```python
mapper = umap.UMAP(random_state=42)
```

```python
embedding = mapper.fit_transform(orig_data)
```
